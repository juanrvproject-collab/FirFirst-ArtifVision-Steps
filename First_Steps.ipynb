```python
import cv2
import time
from IPython.display import display, Image, clear_output

# 1. Iniciar la cámara
captura = cv2.VideoCapture(0)
fps = 30
memoria_video = []

print("Grabando")

# 2. Bucle de GRABACIÓN
for i in range(fps):
    exito, frame = captura.read()
    if exito:
        exito_codificacion, buffer = cv2.imencode('.jpg', frame)
        
        if exito_codificacion:
            memoria_video.append(buffer.tobytes())
    else:
        break


# Liberamos la cámara
captura.release()
print("Grabado")
time.sleep(1) 

# 3. Bucle de REPRODUCCIÓN 
for fotograma_bytes in memoria_video:
    clear_output(wait=True)
    display(Image(data=fotograma_bytes))
    time.sleep(1/fps)

```

```python
import cv2

#MODELO PREENTRENADO PARA DETECTAR ROSTROS
ruta_modelo = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
detector_rostros = cv2.CascadeClassifier(ruta_modelo)

captura = cv2.VideoCapture('Video.mp4') 

print("q para salir")

if not captura.isOpened():
    print("Error: No se pudo abrir el archivo de video. Revisa el nombre y la carpeta.")

while captura.isOpened():
    exito, frame = captura.read()
    
    if not exito:
        print("Terminado.")
        break
        
    # A. Convertir a blanco y negro
    frame_gris = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # B. Detección
    rostros = detector_rostros.detectMultiScale(
        frame_gris, 
        scaleFactor=1.1, 
        minNeighbors=5, 
        minSize=(30, 30)
    )
    
    # C. Dibujar el rectángulo
    for (x, y, w, h) in rostros:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 3)
        cv2.putText(frame, 'Cara', (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
        
    # D. Abrimos la ventana de video
    cv2.imshow("Analizando un video guardado", frame)
    
    # E. Controlamos la velocidad de reproducción. 
    if cv2.waitKey(60) & 0xFF == ord('q'):
        print(" Reproducción interrumpida")
        break

# Liberamos recursos
captura.release()
cv2.destroyAllWindows()
```
